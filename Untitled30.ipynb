```python
import pandas as pd
import random

topics = {
    "gold": [
        "Gold prices rise as investors seek safety",
        "Gold falls due to strong dollar",
        "Gold demand increases amid global uncertainty"
    ],
    "stock": [
        "Stock market crashes amid inflation fears",
        "Equity markets rally on strong earnings",
        "Stocks fall due to recession concerns"
    ],
    "oil": [
        "Crude oil prices surge due to supply disruption",
        "Oil demand weakens in global markets",
        "Oil prices remain stable in trading session"
    ],
    "forex": [
        "Rupee weakens against dollar",
        "Forex market sees high volatility",
        "Dollar strengthens amid economic data"
    ],
    "banking": [
        "Banking stocks surge after profit growth",
        "Bank shares fall due to regulatory concerns",
        "Private banks report strong quarterly results"
    ]
}

sentiments = [
    "boosts investor confidence",
    "creates uncertainty in market",
    "leads to bearish sentiment",
    "drives bullish momentum",
    "impacts global trade outlook"
]

data = []

for i in range(1, 1001):
    topic = random.choice(list(topics.keys()))
    base = random.choice(topics[topic])
    extra = random.choice(sentiments)

    headline = f"{base} and {extra}"

    data.append([i, headline])

df = pd.DataFrame(data, columns=["id", "headline"])

df.to_csv("news_sample.csv", index=False)

print("✅ 1000 news headlines generated successfully!")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, udf
from pyspark.sql.types import DoubleType
from pyspark.ml import Pipeline
from pyspark.ml.feature import (
    Tokenizer,
    StopWordsRemover,
    HashingTF,
    IDF,
    StringIndexer
)
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator


def main():

    # -------------------------------
    # 1. Start Spark Session
    # -------------------------------
    spark = (
        SparkSession.builder
        .appName("MarketSentimentTraining")
        .master("local[*]")
        .getOrCreate()
    )

    spark.sparkContext.setLogLevel("ERROR")
    print("\n🚀 Spark Started...\n")

    # -------------------------------
    # 2. Load Dataset
    # -------------------------------
    df = spark.read.csv(
        "data/labeled_news.csv",
        header=True,
        inferSchema=True
    )

    df = df.dropna()

    print("📄 Dataset Preview:")
    df.show(5, truncate=False)

    # -------------------------------
    # 3. Compute Class Weights
    # -------------------------------
    total_count = df.count()
    class_counts = df.groupBy("sentiment").count().collect()

    class_weight_dict = {}

    for row in class_counts:
        sentiment_label = row["sentiment"]
        count = row["count"]
        class_weight = total_count / (len(class_counts) * count)
        class_weight_dict[sentiment_label] = class_weight

    print("\n📊 Computed Class Weights:")
    for label, weight in class_weight_dict.items():
        print(f"{label}: {weight:.4f}")

    def compute_weight(sentiment):
        return float(class_weight_dict.get(sentiment, 1.0))

    weight_udf = udf(compute_weight, DoubleType())

    df = df.withColumn("classWeight", weight_udf(col("sentiment")))

    print("\n📄 Dataset With Class Weights:")
    df.select("headline", "sentiment", "classWeight").show(5, truncate=False)

    # -------------------------------
    # 4. Convert Labels (text → numeric)
    # -------------------------------
    label_indexer = StringIndexer(
        inputCol="sentiment",
        outputCol="label"
    )

    # -------------------------------
    # 5. Text Processing
    # -------------------------------
    tokenizer = Tokenizer(
        inputCol="headline",
        outputCol="words"
    )

    stopwords = StopWordsRemover(
        inputCol="words",
        outputCol="filtered_words"
    )

    hashing_tf = HashingTF(
        inputCol="filtered_words",
        outputCol="raw_features",
        numFeatures=2000
    )

    idf = IDF(
        inputCol="raw_features",
        outputCol="features"
    )

    # -------------------------------
    # 6. Weighted Logistic Regression
    # -------------------------------
    lr = LogisticRegression(
        featuresCol="features",
        labelCol="label",
        weightCol="classWeight",
        maxIter=30
    )

    # -------------------------------
    # 7. Pipeline
    # -------------------------------
    pipeline = Pipeline(stages=[
        label_indexer,
        tokenizer,
        stopwords,
        hashing_tf,
        idf,
        lr
    ])

    # -------------------------------
    # 8. Train-Test Split
    # -------------------------------
    train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)

    print(f"\n📊 Train size: {train_df.count()}")
    print(f"📊 Test size: {test_df.count()}")

    # -------------------------------
    # 9. Train Model
    # -------------------------------
    print("\n⏳ Training weighted model...\n")
    model = pipeline.fit(train_df)
    print("✅ Weighted Model Trained!\n")

    # -------------------------------
    # 10. Predictions
    # -------------------------------
    predictions = model.transform(test_df)

    print("🔍 Sample Predictions:")
    predictions.select(
        "headline",
        "sentiment",
        "prediction",
        "probability"
    ).show(10, truncate=False)

    # -------------------------------
    # 11. Evaluation
    # -------------------------------
    accuracy_eval = MulticlassClassificationEvaluator(
        labelCol="label",
        predictionCol="prediction",
        metricName="accuracy"
    )

    f1_eval = MulticlassClassificationEvaluator(
        labelCol="label",
        predictionCol="prediction",
        metricName="f1"
    )

    accuracy = accuracy_eval.evaluate(predictions)
    f1_score = f1_eval.evaluate(predictions)

    print(f"\n📈 Accuracy: {accuracy:.4f}")
    print(f"📈 F1 Score: {f1_score:.4f}\n")

    # -------------------------------
    # 12. Save Model
    # -------------------------------
    model_path = "/content/models/sentiment_model"
    # If you want Google Drive instead, use:
    # model_path = "/content/drive/MyDrive/spark_models/sentiment_model"

    model.write().overwrite().save(model_path)

    print(f"💾 Model saved at: {model_path}")

    # -------------------------------
    # 13. Stop Spark
    # -------------------------------
    spark.stop()
    print("\n🛑 Spark Stopped\n")


# -------------------------------
# Run
# -------------------------------
if __name__ == "__main__":
    main()
```

```python
import gradio as gr
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType
from pyspark.ml import PipelineModel

# -------------------------------
# 1. Start Spark Session
# -------------------------------
spark = (
    SparkSession.builder
    .appName("MarketSentimentPredictionApp")
    .master("local[*]")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("ERROR")

# -------------------------------
# 2. Load Trained Model
# -------------------------------
MODEL_PATH = "/content/models/sentiment_model"
# If your model is in Google Drive, use:
# MODEL_PATH = "/content/drive/MyDrive/spark_models/sentiment_model"

model = PipelineModel.load(MODEL_PATH)

# Get label names from StringIndexerModel
label_names = model.stages[0].labels

# -------------------------------
# 3. Topic Detection Function
# -------------------------------
def detect_market_topic(text: str) -> str:
    text = str(text).lower()

    if "gold" in text:
        return "Gold"
    elif "stock" in text or "equity" in text or "share" in text:
        return "Stock"
    elif "oil" in text or "crude" in text:
        return "Oil"
    elif "forex" in text or "rupee" in text or "dollar" in text:
        return "Forex"
    elif "bank" in text or "banking" in text:
        return "Banking"
    return "Other"

# -------------------------------
# 4. Sentiment Badge Formatter
# -------------------------------
def format_sentiment_html(sentiment: str, confidence: float) -> str:
    sentiment_lower = sentiment.lower()

    if sentiment_lower == "positive":
        bg = "#dcfce7"
        text_color = "#166534"
        border = "#86efac"
        emoji = "🟢"
    elif sentiment_lower == "negative":
        bg = "#fee2e2"
        text_color = "#991b1b"
        border = "#fca5a5"
        emoji = "🔴"
    else:
        bg = "#fef3c7"
        text_color = "#92400e"
        border = "#fcd34d"
        emoji = "🟡"

    return f"""
    <div style="
        padding: 10px 14px;
        border-radius: 10px;
        background-color: {bg};
        color: {text_color};
        font-size: 18px;
        font-weight: 700;
        text-align: center;
        border: 1.5px solid {border};
        width: 260px;
        margin: 8px auto 0 auto;
        box-shadow: 0 1px 4px rgba(0,0,0,0.06);
    ">
        <div>{emoji} {sentiment.upper()}</div>
        <div style="font-size: 13px; font-weight: 600; margin-top: 4px;">
            Confidence: {confidence:.2f}%
        </div>
    </div>
    """

# -------------------------------
# 5. Prediction Function
# -------------------------------
def predict_sentiment(headline):
    if not headline or not headline.strip():
        empty_html = """
        <div style="
            padding: 10px 14px;
            border-radius: 10px;
            background-color: #f8fafc;
            color: #475569;
            font-size: 16px;
            font-weight: 700;
            text-align: center;
            border: 1px solid #cbd5e1;
            width: 260px;
            margin: 8px auto 0 auto;
        ">
            Please enter a market news headline
        </div>
        """
        return "", None, empty_html, ""

    schema = StructType([
        StructField("headline", StringType(), True)
    ])

    input_df = spark.createDataFrame([(headline,)], schema=schema)
    prediction_df = model.transform(input_df)

    row = prediction_df.select("prediction", "probability").collect()[0]

    predicted_index = int(row["prediction"])
    predicted_label = label_names[predicted_index]
    probabilities = row["probability"]

    topic = detect_market_topic(headline)
    confidence = float(max(probabilities)) * 100

    sentiment_html = format_sentiment_html(predicted_label, confidence)

    probability_text = "\n".join(
        [f"{label_names[i]}: {float(probabilities[i]) * 100:.2f}%" for i in range(len(label_names))]
    )

    return topic, confidence, sentiment_html, probability_text

# -------------------------------
# 6. Clear Function
# -------------------------------
def clear_all():
    empty_html = """
    <div style="
        padding: 10px 14px;
        border-radius: 10px;
        background-color: #f8fafc;
        color: #475569;
        font-size: 16px;
        font-weight: 700;
        text-align: center;
        border: 1px solid #cbd5e1;
        width: 260px;
        margin: 8px auto 0 auto;
    ">
        Sentiment will appear here
    </div>
    """
    return "", "", None, empty_html, ""

# -------------------------------
# 7. Example Headlines
# -------------------------------
example_inputs = [
    ["Gold prices rise as investors seek safe haven"],
    ["Stock market crashes amid inflation fears"],
    ["Oil prices remain stable in global trade session"],
    ["Rupee weakens against dollar in early market trade"],
    ["Banking stocks surge after strong quarterly results"]
]

# -------------------------------
# 8. Custom CSS
# -------------------------------
custom_css = """
.gradio-container {
    max-width: 1050px !important;
    margin: auto !important;
    background: linear-gradient(180deg, #f8fbff 0%, #f1f5f9 100%);
}

.block-title {
    font-weight: 700;
}

.section-box {
    background: white;
    padding: 18px;
    border-radius: 14px;
    margin-bottom: 16px;
    border: 1px solid #e2e8f0;
    box-shadow: 0 4px 14px rgba(15, 23, 42, 0.06);
}

.main-title {
    text-align: center;
    padding: 10px 0 6px 0;
}

.main-subtitle {
    text-align: center;
    color: #475569;
    margin-bottom: 8px;
}

textarea, input {
    font-size: 15px !important;
}

button {
    border-radius: 10px !important;
}

.prob-box textarea {
    font-family: monospace !important;
}
"""

# -------------------------------
# 9. Gradio Interface
# -------------------------------
with gr.Blocks(css=custom_css, title="Market Sentiment Analyzer") as demo:

    # Header
    gr.Markdown("""
    <div class="main-title">
        <h1>📊 Spark-Trained Market Sentiment Analyzer</h1>
    </div>
    <div class="main-subtitle">
        Analyze financial headlines using your trained Apache Spark model
    </div>
    """)

    with gr.Tab("Input"):
        with gr.Group(elem_classes="section-box"):
            gr.Markdown("### 📰 Enter News Headline")
            news_input = gr.Textbox(
                label="Market News Input",
                placeholder="Example: Rupee weakens against dollar in early market trade",
                lines=3
            )

            with gr.Row():
                predict_btn = gr.Button("🔍 Analyze Sentiment", variant="primary")
                clear_btn = gr.Button("🧹 Clear")

    with gr.Tab("Result"):
        with gr.Group(elem_classes="section-box"):
            gr.Markdown("### 📌 Prediction Summary")
            with gr.Row():
                topic_output = gr.Textbox(label="Detected Topic")
                confidence_output = gr.Number(label="Confidence (%)", precision=2)

            sentiment_output = gr.HTML(
                value="""
                <div style="
                    padding: 10px 14px;
                    border-radius: 10px;
                    background-color: #f8fafc;
                    color: #475569;
                    font-size: 16px;
                    font-weight: 700;
                    text-align: center;
                    border: 1px solid #cbd5e1;
                    width: 260px;
                    margin: 8px auto 0 auto;
                ">
                    Sentiment will appear here
                </div>
                """,
                label="Highlighted Sentiment"
            )

    with gr.Tab("Probabilities"):
        with gr.Group(elem_classes="section-box"):
            gr.Markdown("### 📈 Class Probability Distribution")
            probability_output = gr.Textbox(
                label="Model Probabilities",
                lines=6,
                elem_classes="prob-box"
            )

    with gr.Tab("Examples"):
        with gr.Group(elem_classes="section-box"):
            gr.Markdown("### 💡 Sample Headlines")
            gr.Examples(
                examples=example_inputs,
                inputs=news_input
            )

    # Button actions
    predict_btn.click(
        fn=predict_sentiment,
        inputs=news_input,
        outputs=[topic_output, confidence_output, sentiment_output, probability_output]
    )

    clear_btn.click(
        fn=clear_all,
        inputs=[],
        outputs=[news_input, topic_output, confidence_output, sentiment_output, probability_output]
    )

demo.launch(debug=True)
```
